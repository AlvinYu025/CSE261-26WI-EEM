```python
import pandas as pd
from pathlib import Path

# Define cities/datasets
cities = ['am', 'mo', 'ny']

for city in cities:
    print(f"Processing {city}...")
    
    # Read mismatch score CSV (adjust path/filename as needed)
    mismatch_csv_path = f"../../mismatch-score-csv/{city}/llm_mismatch_score.csv"  
    mismatch_df = pd.read_csv(mismatch_csv_path)
    
    # Read original reviews CSV
    reviews_csv_path = f"../../airbnb-{city}-data/reviews.csv"
    reviews_df = pd.read_csv(reviews_csv_path, low_memory=False)
    
    # Select only the columns we need from reviews_df (id and date)
    date_columns = reviews_df[['id', 'date']] 
    
    # Merge on review id 
    merged_df = mismatch_df.merge(date_columns, left_on='review_id', right_on='id', how='left')
    
    # Save the result
    output_path = f"../../mismatch-score-csv/{city}/llm_mismatch_score_with_date.csv"
    merged_df.to_csv(output_path, index=False)
    
    print(f"Saved to {output_path}")
    print(f"Rows: {len(merged_df)}, Columns: {merged_df.columns.tolist()}\n")
```
